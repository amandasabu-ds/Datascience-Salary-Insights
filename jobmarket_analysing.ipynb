```python
# Import pandas library
import pandas as pd

# Load the CSV file
df = pd.read_csv("ds_salaries.csv")

# Show the first 5 rows
df.head()
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt
```

```python
df['employment_type'].value_counts().plot(kind='bar')
plt.show()
```

```python
df.groupby('employment_type')['salary_in_usd'].mean().plot(kind='bar', color='skyblue')
plt.title("Average Salary by Employment Type")
plt.xlabel("Employment Type")
plt.ylabel("Average Salary (USD)")
plt.show()
```

```python
df['employment_type'].value_counts().plot(kind='bar', color='orange')
plt.title("Number of Employees by Employment Type")
plt.xlabel("Employment Type")
plt.ylabel("Count")
plt.show()
```

```python
df.groupby('experience_level')['salary_in_usd'].mean().plot(kind='bar', color='green')
plt.title("Average Salary by Experience Level")
plt.xlabel("Experience Level")
plt.ylabel("Average Salary (USD)")
plt.show()
```

```python
df['experience_level'].value_counts().plot(kind='bar', color='purple')
plt.title("Number of Employees by Experience Level")
plt.xlabel("Experience Level")
plt.ylabel("Count")
plt.show()
```

```python
df.groupby('job_title')['salary_in_usd'].mean().sort_values(ascending=False).head(10).plot(kind='barh', color='teal')
plt.title("Top 10 Highest Paying Job Roles")
plt.xlabel("Average Salary (USD)")
plt.ylabel("Job Title")
plt.show()
```

```python
df['job_title'].value_counts().head(10).plot(kind='bar', color='pink')
plt.title("Top 10 Most Common Job Roles")
plt.xlabel("Job Title")
plt.ylabel("Count")
plt.show()
```

# Job Market Analysis Project

## Dataset Overview
- *Dataset:* ds_salaries.csv  
- *Columns:* job_title, employment_type, experience_level, salary_in_usd, work_year, company_location, remote_ratio, company_size, etc.  
- *Goal:* Analyze global data science job trends, salaries, and employment types.

---

## Key Insights

### 1 Employment Type
- Full-Time (FT) employees dominate the workforce.  
- Contractors (CT) and Freelancers (FL) earn competitive average salaries.  
- Part-Time (PT) roles are fewer but still significant.  

### 2 Experience Level
- Higher experience correlates with higher salaries:
  - EN = Entry Level  lower average salary  
  - MI = Mid Level  moderate salary  
  - SE = Senior  higher salary  
  - EX = Executive  highest salaries  

### 3 Job Roles
- Top paying roles: applied scientist,datascience manager,machine learning software engineer 
- Most common roles: data engineer,data scientist,data analyst  

### 4 Salary Analysis
- Average salary varies significantly based on employment type, experience level, and job title.  
- Insights can guide job seekers, HR managers, and researchers.

---

## Conclusion
This project provides a clear view of the global data science job market.  
By analyzing salaries, employment types, experience levels, and job roles, we can understand:
- Which roles pay the most  
- Which employment types are most common  
- How experience affects compensation
